```python
#통합 데이터
whole_data=pd.concat([safe, youth_institution,youth_population, smoke, run_away, smartphone, stress], axis=1)
whole_data.columns=['보호소년(명)', '복지시설(개)', '청소년 인구수(명)', '흡연(명)', '음주(명)', '가출경험(%)', '스마트폰 과의존(%)', '스트레스 및 우울감(%)']
whole_data=whole_data.head(96)
whole_data
```

## 데이터 표준화

```python
#피처 스케일링
pd.options.mode.chained_assignment = None
def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x: (x-series_mean)/series_std)
    return df
```

```python
scale_columns = ['복지시설(개)', '청소년 인구수(명)', '가출경험(%)', '스트레스 및 우울감(%)']
whole_df = whole_data[['청소년 인구수(명)', '흡연(명)', '음주(명)', '복지시설(개)', '보호소년(명)', '가출경험(%)',  '스트레스 및 우울감(%)']]
X = whole_df.loc[ : , whole_df.columns != '보호소년(명)']
whole_data1= standard_scaling(whole_df , scale_columns)
whole_data1
```

## 데이터 예측 

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
#청소년 범죄율
X = whole_data1[whole_data1.columns.difference(['보호소년(명)'])]
y = whole_data1['보호소년(명)']
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state=20)
```

```python
lr1 = linear_model.LinearRegression()
model1 = lr1.fit(X_train , y_train)
```

```python
print(lr1.coef_)
```

```python
# R2 score
print(model1.score(X_train , y_train))
print(model1.score(X_test , y_test))
```

```python
# RMSE score
y_predictions = lr1.predict(X_train)
print(sqrt(mean_squared_error(y_train, y_predictions))) 
y_predictions = lr1.predict(X_test)
print(sqrt(mean_squared_error(y_test, y_predictions)))
```
